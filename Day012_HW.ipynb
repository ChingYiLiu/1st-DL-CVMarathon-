### ① 取出今天所有的發文

```python
import requests
from bs4 import BeautifulSoup
import re

url = 'https://www.ettoday.net/news/news-list.htm'
r = requests.get(url)

soup = BeautifulSoup(r.text, "html5lib")
```

```python
print(soup.find(class_="part_list_2"))
```

```python
regex = re.compile('tag c_*')

d = soup.find(class_="part_list_2").find_all('h3')[0]
print(d)
print(d.find_all("em", {"class": regex})[0].text)
print(d.find_all('a')[0].text)
```

```python
for d in soup.find(class_="part_list_2").find_all('h3'):
    print(d.find(class_="date").text, d.find_all("em", {"class": regex})[0].text ,d.find_all('a')[0].text)
```

### ② 如果想要依照類別分類，怎麼儲存會比較好？

```python
import pandas as pd
dataset = pd.DataFrame(columns = ['Date', 'Catogery', 'Title'])
```

```python
print(dataset)
```

```python

for d in soup.find(class_="part_list_2").find_all('h3'):
   # print(d[1])
    date = d.find(class_="date").text
    cat = d.find_all("em", {"class": regex})[0].text
    title = d.find_all('a')[0].text
    dataset = dataset.append({'Date': date, 'Catogery': cat, 'Title': title}, ignore_index=True)
```

```python
print(dataset)
```

### ③ 哪一個類別的文章最多

```python
dataset.groupby('Catogery').count().sort_values('Title', ascending=False ).iloc[0]
```
