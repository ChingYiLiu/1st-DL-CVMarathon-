### 尋找一個網站是有作 Header 檢查的（Hint: 有加跟沒有加回傳結果不一樣）

```python
import requests
```

```python
r = requests.get('https://www.bloomberg.com/asia')
response = r.text

print(response)
```

### 用上述網站說明該如何判斷 Header 中必須加上哪些資料才會正確


```python
header = {
    'path': '/asia',
    'user-agent': 'Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/87.0.4280.88 Safari/537.36'
}
r = requests.get('https://www.bloomberg.com/asia', headers = header)
response = r.text

print(response)
```
