    2. 取出第一筆與最後一筆回答的時間


### 1. 取出知乎問題發問時間

```python
# 1. 取出知乎問題發問時間
import requests
header = {'user-agent': 'my-app/0.0.1'}
r = requests.get('https://www.zhihu.com/api/v4/questions/55493026/answers', headers =header)
response = r.text
```

```python
import json
from pprint import pprint
data = json.loads(response)

for d in data['data']:
    pprint(d)
```

### 2. 取出第一筆與最後一筆回答的時間

```python
# 2. 取出第一筆與最後一筆回答的時間
for item in range(1,len(data['data'])):
    x = data['data'][item]['created_time']
    print('%s  %s' %(item, x))
```
