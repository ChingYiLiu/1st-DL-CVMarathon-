## 作業目標

1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

```python
import requests
import json
from pprint import pprint

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')

rd = json.loads(r.text)
#json.loads(r.text)
pprint(rd)
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

print('資料筆數：  %s' %(len(rd)))

rd_dict = dict(enumerate(rd, start = 1)) #指定從1開始編碼(python預設是0)
dict_keys = list(rd_dict[1].keys())
print('資料欄位： %s' %(dict_keys))
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
import pandas as pd

target = pd.DataFrame()

for item in rd:
    print(item['title'])
    print(item['createdAt'])
    print(item['commentCount'])
    print(item['likeCount'])
 
```

```python
# 3. 計算熱門/非熱門文章的「平均留言人數」與「平均按讚人數」
## 熱門
commentcount = 0
likecount = 0
ttle = len(rd)
for item in rd:
    commentcount += item['commentCount']
    likecount += item['likeCount']

avg_comment = commentcount/ttle
avg_like = likecount/ttle
    
    
## 非熱門
r2 = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=false')
rd2 = json.loads(r2.text)

commentcount2 = 0
likecount2 = 0
ttle2 = len(rd2)
for item in rd2:
    commentcount2 += item['commentCount']
    likecount2 += item['likeCount']

avg_comment2 = commentcount2/ttle2
avg_like2 = likecount2/ttle2
    
```

```python
print('[熱門文章]')
print('平均留言人數： %.2f \n平均按讚人數： %.2f' %(avg_comment, avg_like))
print(' ')
print('[非熱門文章]')
print('平均留言人數： %.2f \n平均按讚人數： %.2f' %(avg_comment2, avg_like2))
```
