```python
import requests
import json
from pprint import pprint

r = requests.get('https://www.dcard.tw/_api/forums/pet/posts?popular=true')

rd = json.loads(r.text)
#json.loads(r.text)
pprint(rd)
```

```python
# 1. 這個 API 一次會回傳幾筆資料？每一筆資料包含哪些欄位？

print('資料筆數：  %s' %(len(rd)))

rd_dict = dict(enumerate(rd, start = 1)) #指定從1開始編碼(python預設是0)
dict_keys = list(rd_dict[1].keys())
print('資料欄位： %s' %(dict_keys))
```

```python
# 2. 取出每一筆資料的「標題」、「貼文時間」、「留言人數」、「按讚人數」
import pandas as pd

target = pd.DataFrame()

for nums in rd_dict:
    target_dict = pd.Series(rd_dict[nums][2])
    target_dict2 = pd.DataFrame(target_dict).transpose()
    target = target.append(target_dict2)
#target_dict = pd.Series(rd_dict[2])
#target_dict2 = pd.DataFrame(target_dict).transpose()
```
